#### Question 11 - How many customers downgraded from a pro monthly to a basic monthly plan in 2020?

```python
%%sql
with stag as (
select *
, case when date(s2.start_date) < date(s1.start_date) then 'downgrade' 
    else 'upgrade' end plan_switch
from subscription s1
join subscription s2
on s1.customer_id = s2.customer_id
and strftime('%Y',date(s1.start_date)) = '2020'
where s1.plan_id = 1
and s2.plan_id = 2
)

select count(*) cust_count
from stag
where plan_switch = 'downgrade'
```

- There are no customers who downgraded from a `pro monthly` to a `basic monthly` plan.
